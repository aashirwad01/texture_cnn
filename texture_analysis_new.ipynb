```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.utils import to_categorical
from keras.preprocessing import image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from tqdm import tqdm
import cnn_utils as utils

from keras.layers import Conv2D, MaxPooling2D,BatchNormalization
```

```python
y_test = to_categorical(y_test)
y_train = to_categorical(y_train)
```

```python
model = Sequential()
model.add(Conv2D(64, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=(200,200,3)))

model.add(Conv2D(64, (3,3), 
                activation='relu'))
# kernel regulization in convolutional neural networks is a way reduce again the datasets no of parameters by taking a sparse matrix and multiplying with pixel matrix of the image   
# model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))
#In Max pooling we choose the maximum value within a matrix. The size of the matrix could be 2 X 2 or 3 X 3 also. Here is an image showing the max pooling of the reference image of 3
#It is basically used to reduce the size of the image because the larger number of pixels contribute to more parameters which can involve large chunks of data. Thus we need less parameters such that a CNN can still identify the imag
# model.add(Dropout(0.2))

model.add(Flatten())
model.add(Dense(256, activation='relu'))
# this dense layer seems to be second  last layer that means layer before the output layer having 128 neurons 
model.add(Dropout(0.5))
# dropout is added to the hidden layer randomly drops Dropout is a technique where randomly selected neurons are ignored during training. They are “dropped-out” randomly.
model.add(Dense(62, activation='softmax'))
# softmax activation is used in the output layer at the end usually before coverting data to output classes
```

```python
model.summary()
```

```python
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),metrics=['accuracy'])
```

```python
history=model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
```

```python
score = model.evaluate(x_test, y_test, verbose=1)
```
